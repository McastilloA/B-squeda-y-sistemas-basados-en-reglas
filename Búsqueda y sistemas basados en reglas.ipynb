```python
import osmnx as ox
```

```python
nombre_ciudad = "Ciudad Jardín, Córdoba, Spain"
grafico_ciudad = ox.graph_from_place(nombre_ciudad, network_type="drive")
```

```python
fig, ax = ox.plot_graph(grafico_ciudad, node_color="y")
```

```python
ox.plot_graph_folium(grafico_ciudad, popup_attribute='name', color='blue', weight=3, opacity=0.9)
```

```python
ox.save_graph_geopackage(grafico_ciudad, filepath="mosquera.png")
```

```python
origen = ox.distance.nearest_nodes(grafico_ciudad, X=-4.7943378, Y=37.8831953)
destino = ox.distance.nearest_nodes(grafico_ciudad, X=-4.7870238, Y=37.8831438)
```

```python
ruta = ox.shortest_path(grafico_ciudad, origen, destino, weight="length")
fig, ax = ox.plot_graph_route(grafico_ciudad, ruta, node_size=0)
```
